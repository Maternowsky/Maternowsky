# **Exploring Unsupervised learning techniques. Clustering is finding natural grouping in data so that items in same cluster are more similar to each other than those of different clusters**

## **Grouping objects using k-means. this algorithm belongs to the category of prototype-based clustering. Have to define number of clusters and is good at identifying clusters with spherical shape**

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples = 150,
                  n_features = 2,
                  centers = 3,
                  cluster_std = 0.5,
                  shuffle = True,
                  random_state = 0)
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1],
            c = 'white',
            marker = 'o',
            edgecolor = 'black',
            s = 50)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid()
plt.tight_layout()
plt.show()
```

### **Commonly used distance for clustering examples with continuous features is (squared Euclidean distance). Based on this Euclidean metric we can define k-means algorithm as iterative approach for minimizing the within-cluster (sum of squared errors(SSE)) also called cluster inertia**

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3,
            init = 'random',
            n_init = 10,
            max_iter = 300,
            tol= 1e-04,
            random_state=0)
y_km = km.fit_predict(X)
```

### **viewing the clusters predicted. Cluster centroids stored in cluster_centers_ attribute**

```python
plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s = 50, c = 'lightgreen',
            marker = 's', edgecolor = 'black',
            label = 'Cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s=50, c = 'orange',
            marker = 'o', edgecolor = 'black',
            label = 'Cluster 2')
plt.scatter(X[y_km == 2, 0],
            X[y_km ==2, 1],
            s = 50, c = 'lightblue',
            marker = 'v', edgecolor = 'black',
            label = 'Cluster 3')
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s = 250, marker = '*',
            c = 'red', edgecolor = 'black',
            label = 'Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend(scatterpoints = 1)
plt.grid()
plt.tight_layout()
plt.show()
```

# **Smarter way of placing initial cluster centroids using k-means++. to use kmean++ in scikit learn just set init='kmeans++'. this is also the default setting**

## **Checking quality of clustering using within-cluster SSE(distortion)**

```python
print(f'Distortion: {km.inertia_:.2f}')
```

## **Using graphical tool, Elbow Method, to estimate optimal number of clusters, k**

```python
distortions = []
for i in range(1, 11):
  km = KMeans(n_clusters = i,
              init = 'k-means++',
              n_init = 10,
              max_iter = 300, 
              random_state = 0)
  km.fit(X)
  distortions.append(km.inertia_)

plt.plot(range(1,11), distortions, marker = 'o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
plt.show()
```

### **We can see the 'Elbow' is located at k = 3 so this would be a good value for k**

## **Quantifying the quality of clustering via silhouette plots**

```python
km = KMeans(n_clusters = 3,
            init = 'k-means++',
            n_init= 10,
            max_iter = 300,
            tol= 1e-04,
            random_state = 0)
y_km = km.fit_predict(X)
import numpy as np
from matplotlib import cm
from sklearn.metrics import silhouette_samples
cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(X, y_km, metric = 'euclidean')
y_ax_lower, y_ax_upper = 0, 0
yticks = []


for i, c in enumerate(cluster_labels):
  c_silhouette_vals = silhouette_vals[y_km == c]
  c_silhouette_vals.sort()
  y_ax_upper += len(c_silhouette_vals)
  color = cm.jet(float(i)/n_clusters)
  plt.barh(range(y_ax_lower, y_ax_upper),
           c_silhouette_vals,
           height = 1.0,
           edgecolor = 'none',
           color = color)
  yticks.append((y_ax_lower + y_ax_upper)/2.)
  y_ax_lower += len(c_silhouette_vals)

silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg,
            color = 'red',
            linestyle = '--')
plt.yticks(yticks, cluster_labels +1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')
plt.tight_layout()
plt.show()
```

### **Looking at BAD clustering. using two centroids**

```python
km = KMeans(n_clusters=2,
            init = 'k-means++',
            n_init = 10,
            max_iter = 300,
            tol=1e-04,
            random_state=0)
y_km = km.fit_predict(X)
plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s = 50, c = 'lightgreen',
            edgecolor = 'black',
            marker = 's',
            label = 'Cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s = 50, c = 'orange',
            edgecolor = 'black',
            marker = 'o',
            label = 'Cluster 2')
plt.scatter(km.cluster_centers_[:,0],
            km.cluster_centers_[:,1],
            s = 250, marker = '*',
            c = 'red',
            label = 'centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

### **Creating Silhouette plot to evaluate results**

```python
cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(X, y_km, metric='euclidean')
y_ax_lower, y_ax_upper = 0, 0
yticks = []
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(float(i) / n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, 
             edgecolor='none', color=color)

    yticks.append((y_ax_lower + y_ax_upper) / 2.)
    y_ax_lower += len(c_silhouette_vals)
    
silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color="red", linestyle="--") 

plt.yticks(yticks, cluster_labels + 1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')

plt.tight_layout()
#plt.savefig('figures/10_06.png', dpi=300)
plt.show()
```

# **Hierarchical Clustering- one advantage is we can plot dendrograms**

## **Agglomerative hierarchical clustering- two standard algorithms are single linkage and complete linkage. in next section we will focus on complete linkage**

```python
import pandas as pd
import numpy as np
np.random.seed(123)
variables = ['X', 'Y', 'Z']
labels = ['ID_0', 'ID_1', 'ID_2', 'ID_3', 'ID_4']
X = np.random.random_sample([5,3])*10

df = pd.DataFrame(X, columns = variables, index = labels)
df
```

```python
from scipy.spatial.distance import pdist, squareform
row_dist = pd.DataFrame(squareform(pdist(df, metric='euclidean')),
                        columns=labels,
                        index = labels)
row_dist
```

### **In preceding code we calculated the Euclidean distance between each pair of input examples based on features X, Y, Z. We provided condensed distance matrix(with pdist) as input to (squareform) function to create symmetrical matrix**

```python
# looking at 'linkage' function documentation

from scipy.cluster.hierarchy import linkage
help(linkage)
```

### **Incorrect Approach (squareform)**

```python
row_cluster = linkage(row_dist,
                      method = 'complete',
                      metric = 'euclidean')
```

### **Correct Approaches**

```python
row_clusters = linkage(pdist(df, metric = 'euclidean'),
                       method = 'complete')

row_clusters = linkage(df.values, 
                       method = 'complete',
                       metric = 'euclidean')
```

### **To take a closer look at the clustering results, we can turn those results into a pandas dataframe**

```python
pd.DataFrame(row_clusters,
             columns = ['row label 1',
                        'row label 2',
                        'distance',
                        'no. of items in clust.'],
             index = [f'cluster {(i+1)}' for i in range(row_clusters.shape[0])])
```

## **Viewing results in form of Dendrogram**

```python
from scipy.cluster.hierarchy import dendrogram
#make dendrogram black (part 1/2)
#from scipy.cluster.hierarchy import set_link_color_palette
# set_link_color_palette(['black'])
row_dendr = dendrogram(row_clusters,
                       labels = labels,
                       # make dendogram black (part 2/2)
                       # color_threshold = np.inf
                        )
plt.tight_layout()
plt.ylabel('Euclidean distance')
plt.show()
```

### **How to attach dendrogram to a heat map**

```python
# Step 1
# Create new figure object and define x axis position, y axis position,
# width, height via the add_axes attribute

fig = plt.figure(figsize = (8, 8), facecolor='white')
axd = fig.add_axes([0.09, 0.1, 0.2, 0.6])
row_dendr = dendrogram(row_clusters,
                       orientation = 'left')
# for matplotlib < v1.5.1 please use orientation = 'right'

# reorder the data in Dataframe according to clustering labels

df_rowclust = df.iloc[row_dendr['leaves'][::-1]]

# construct heat map for the reordered DataFrame and position it next to dendrogram

axm = fig.add_axes([0.23, 0.1, 0.6, 0.6])
cax = axm.matshow(df_rowclust,
                  interpolation = 'nearest',
                  cmap = 'hot_r')

# modify aesthetics of dendrogram by removing the axis ticks and 
# hiding the axis spines, add a color bar and assign the feature
# and data record names to the x and y axis tick labels

axd.set_xticks([])
axd.set_yticks([])
for i in axd.spines.values():
  i.set_visible(False)
fig.colorbar(cax)
axm.set_xticklabels([''] + list(df_rowclust.columns))
axm.set_yticklabels([''] + list(df_rowclust.index))
plt.show()
```

# **DBSCAN - density-based spatial clustering of applications with noise**

## **Label assigned using following criteria- Core point - at least a specified number of neighboring points fall within the specified radius, Border point- point that has fewer neighbors than MinPts within epsilon, but lies within radius of core point. Noise points- all other points**

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples = 200,
                  noise = 0.05,
                  random_state = 0)
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.tight_layout()
plt.show()
```

### **Seeing if k-means can identify the half-moons**

```python
f, (ax1, ax2) = plt.subplots(1,2,figsize=(8,3))
km = KMeans(n_clusters=2, random_state=0)
y_km = km.fit_predict(X)
ax1.scatter(X[y_km ==0, 0],
            X[y_km == 0, 1],
            c = 'lightblue',
            edgecolor = 'black',
            marker = 'o',
            s = 40,
            label = 'cluster 1')
ax1.scatter(X[y_km ==1, 0],
            X[y_km ==1, 1],
            c = 'red',
            edgecolor = 'black',
            marker = 's',
            s = 40,
            label = 'cluster 2')
ax1.set_title('K-means clustering')
ax1.set_xlabel('Feature 1')
ax1.set_ylabel('Feature 2')
ac = AgglomerativeClustering(n_clusters=2,
                             metric = 'euclidean',
                             linkage = 'complete')
y_ac = ac.fit_predict(X)
ax2.scatter(X[y_ac == 0, 0],
            X[y_ac == 0, 1],
            c = 'lightblue',
            edgecolor = 'black',
            marker = 'o',
            s = 40,
            label = 'Cluster 1')
ax2.scatter(X[y_ac == 1, 0],
            X[y_ac == 1, 1],
            c = 'red',
            edgecolor = 'black',
            marker = 's',
            s = 40,
            label = 'Cluster 2')
ax2.set_title('Agglomerative clustering')
ax2.set_xlabel('Feature 1')
ax2.set_ylabel('Feature 2')
plt.legend()
plt.tight_layout()
plt.show()
```

### **Trying DBSCAN algorithm**

```python
from sklearn.cluster import DBSCAN
db = DBSCAN (eps = 0.2,
             min_samples = 5,
             metric = 'euclidean')
y_db = db.fit_predict(X)
plt.scatter(X[y_db == 0, 0],
            X[y_db == 0, 1],
            c = 'lightblue',
            edgecolor = 'black',
            marker = 'o',
            s=40,
            label = 'Cluster 1')
plt.scatter(X[y_db == 1, 0],
            X[y_db == 1, 1],
            c = 'red',
            edgecolor = 'black',
            marker = 's',
            s = 40,
            label = 'Cluster 2')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.tight_layout()
plt.show()
```
